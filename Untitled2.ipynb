```python
from rdkit import Chem, DataStructs
from rdkit.Chem import Draw, AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit import rdBase
rdBase.DisableLog('rdApp.error')
rdBase.DisableLog('rdApp.info')  

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('comparetwosmiles.csv', delimiter='\t')

```

```python
# Left SMILEs
for i in range(0, len(df)):
    smiles = Chem.MolFromSmiles(df['canonical_smiles'][i])
    PILimg = Draw.MolToImage(smiles)
    print(df['canonical_smiles'][i])
    display(PILimg)
```

```python
#right smiles
for i in range(0, len(df)):
    smiles = Chem.MolFromSmiles(df['canonical_smiles.1'][i])
    PILimg = Draw.MolToImage(smiles)
    print(df['canonical_smiles.1'][i])
    display(PILimg)
```
